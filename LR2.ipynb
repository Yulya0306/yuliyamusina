```python
# Задание 1
import pandas as pd
```

```python
mov=pd.read_csv('movies.csv')
mov[['movieId','title']].head(1000)
```

```python
rat=pd.read_csv('ratings.csv')
rat=rat[['movieId','rating']].head(1000)
```

```python
joined = rat.merge(mov, on='movieId', how='left')
joined[['movieId','rating','title']].head(1000)
```

```python
max_=rat['rating'].max()
filtered_joined=joined[(joined ['rating']==max_)]
film=filtered_joined.groupby(['movieId' , 'title', 'rating'])['movieId'].count()
```

```python
film.sort_values(ascending=False).head()
```
